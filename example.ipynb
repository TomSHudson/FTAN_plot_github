# Example of using FTAN_plot module:

Summary:

The FTAN plot effectively switches the resolution between high resolution in time (for the spectrogram plot) and high resolution in frequency (FTAN plot). Narrowing the centre frequency step increases the apparent resolution, as does increasing the number of centre frequency steps.

## 1. Get data:

```python
# Import neccessary modules:
import FTAN_plot
# Modules used only in this example (not in FTAN_plot):
import obspy
import matplotlib.pyplot as plt

# Import data for example:
st = obspy.read("S_waveform_E_2014_180_2037_event.m")
data = st[0].data # Data associated with trace of real waveform observed
```

## 2. Plot seismic waveform:

```python
# Plot seismic signal:
st.plot()
```

## Plot FTAN style plot:

```python
# Plot FTAN style plot for signal:
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
FTAN_plot.ftan_plot(data, samp_rate=500.0, centre_freq_range=[4.0,50.0], centre_freq_range_step=0.25, band_width_gau_filter=[1.00], axes=ax)
```

## Plot spectrogram for comparison:

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
st[0].spectrogram(axes=ax)
ax.set_ylim([1.,50.])
ax.set_xlim([0.125,1.825])
plt.show()
```
